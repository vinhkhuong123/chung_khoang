```python
from bs4 import BeautifulSoup
import requests as rq
```

```python
# tạo 1 biến url để lưu link web
url = "https://cafef.vn/thi-truong-chung-khoan.chn"
```

```python
# Gửi yêu cầu GET đến trang web và lấy nội dung
content = rq.get(url).content
print(content)
```

```python
# Sử dụng BeautifulSoup để phân tích cú pháp HTML, làm đẹp với kiểu html.parser
soup = BeautifulSoup(content, "html.parser")
soup
```

```python
# Lưu dữ liệu vào tệp tin
with open("data.html", "wb") as file:
    file.write(content)
```

```python
# tạo 1 biến body_part để lưu dữ liệu trong thẻ <body>
body_part = soup.find('body')
body_part
```

```python
# tạo 1 biến d_main dùng findAll để tìm kiếm thẻ chứa thông tin cần tìm
d_main = body_part.findAll('div', {'class':'tlitem box-category-item'})
```

```python
# in thử thông tin đầu
print(d_main[0])
```

# từ thông tin trên ta đã phân tích được các nội dung sau
* thẻ h3 là tên tiêu đề bài viết.    
* thẻ span là thẻ chứa thông tin thời gian
* thẻ p class="sapo box-category-sapo là thẻ chưa nôi dụng chính

```python
# tạo 1 mảng cho tiêu đề
ten_tieude = []
```

```python
# tạo 1 vòng lặp qua mỗi 1 nội dung trong d_main 
# sử dụng find() để lấy nội dung của thẻ 
# text.strip() để lấy text trong thẻ bỏ qua các khoảng trắng ở đầu và cuối.
# append() để thêm vào mảng
for i in d_main:
    ten_tieude.append(i.find('h3').text.strip())
```

```python
# in thử tên đầu
print(ten_tieude[0])
```

```python
# kiểm tra độ dài mảng
len(ten_tieude)
```

```python
# tạo 1 mảng cho thời gian
thoi_gian = []
```

```python
for i in d_main:
    thoi_gian.append(i.find('span').text.strip())
```

```python
thoi_gian[0]
```

```python
len(thoi_gian)
```

```python
# tạo 1 mảng cho nội dung
noi_dung = []
```

```python
for i in d_main:
    noi_dung.append(i.find('p',{'class':'sapo box-category-sapo'}).text.strip())
```

```python
noi_dung[0]
```

```python
len(noi_dung)
```

```python
# khai báo thư viện pandas
import pandas as pd
```

```python
# tạo 1 dataFrame từ các mảng trên
data = {'time':thoi_gian, 'title':ten_tieude, 'main':noi_dung}

df = pd.DataFrame(data)
df
```

```python
# in ra 1 file csv
df.to_csv('file_chung_khoang.csv', index=False)
```
